```python
from math import pi
from numpy import *
from scipy.stats import norm
from scipy.constants import *
from ipywidgets import *
from matplotlib import pyplot as plt
from matplotlib import rcParams as MpRcParams
MpRcParams['mathtext.fontset'] = 'cm'
MpRcParams['mathtext.rm'] = 'serif'
MpRcParams['mathtext.rm'] = 'serif'
%matplotlib inline
```

# Slike za predavanja

## Spektralna učinkovitost prenosa

```python

CzB=logspace(-2,1,1000)
Ebn0=(2**CzB-1)/CzB
Ebn0_dB=10*log10(Ebn0)

#Risanje
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(Ebn0_dB,CzB)
ax.axvline(0, color='black', lw=1)
ax.axvline(-1.6, color='red', lw=1,dashes=(5,5))
ax.axhline(1, color='black', lw=1)
plt.ylabel(r'$\frac{C}{B}\rm{[sh/s/Hz]}$', fontsize=28)
plt.xlabel(r'$\frac{E_b}{n_0}\rm{[dB]}$', fontsize=28)
plt.yscale("log")
plt.grid(which="both")
ax.set_frame_on(False)
plt.rc('ytick', labelsize=16)
plt.rc('xtick', labelsize=16) 
```
