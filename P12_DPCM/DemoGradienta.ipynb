# Demo 1D gradienta

```python
a1=linspace(-1,3,100)
sigmad2= (a1-1)**2+1
gradSigmad2=sigmad2[1:]-sigmad2[:-1]

#Risanje
fig, ax= plt.subplots(figsize=(6,6))
ax.plot(a1,sigmad2,label="$\sigma_d^2$")
ax.plot(a1[1:],gradSigmad2*10,'r',label=r'$\frac{d\sigma_d^2}{da_1}$')
ax.axhline(0, color='black', lw=1)
ax.axvline(0, color='black', lw=1)
ax.legend(fontsize=20)
plt.ylabel(r'$\sigma_d^2$ ,  $\frac{d\sigma_d^2}{da_1}$', fontsize=28)
plt.xlabel(r'$a_1$', fontsize=28)
ax.set_frame_on(False)
plt.rc('ytick', labelsize=16)
plt.rc('xtick', labelsize=16) 

```

# Demo 2D gradienta

```python
a1=linspace(-1,3,100)
a2=linspace(-1,3,100)
#sigmad2= (a1-1)**2+1+(a2-1)**2

X,Y = meshgrid(a1,a2)
sigmad2= (X-1)**2+(Y-1)**2



fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, sigmad2,cmap='plasma', edgecolor='none',alpha=0.7)
ax.set_xlabel(r'$a_1$',fontsize=18)
ax.set_ylabel(r'$a_2$',fontsize=18)
ax.set_zlabel(r'$\sigma_d^2$',fontsize=18)
plt.show()        
```
