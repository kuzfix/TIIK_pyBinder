### Jupyterlab magic
Spodnja celica uvozi vse knjižnice (mogoče tudi kakšno preveč) in pripravi nastavitve, ki jih potrebujemo kasneje za računanje in risanje. Poženi pred čemerkoli drugim

```python
from math import pi
from numpy import *
from scipy.stats import norm
from scipy.constants import *
from ipywidgets import *
from matplotlib import pyplot as plt
from matplotlib import rcParams as MpRcParams
MpRcParams['mathtext.fontset'] = 'cm'
MpRcParams['mathtext.rm'] = 'serif'
MpRcParams['mathtext.rm'] = 'serif'
%matplotlib inline
```

# Porazdelitev pogojne gostote verjetnosti kanala

### Primer 1:

```python
N=0.04
sigma_n=sqrt(N)
X=array([-1, 0, 1])

#Izracunaj tabelo vrednosti za risanje grafa
fig, ax = plt.subplots(figsize=(12,4))
y=linspace(-2,2,1000)
for x in X:
    pnor=norm(scale=sigma_n,loc=x)
    pypx=pnor.pdf(y)
    ax.plot(y,pypx)
    ax.axvline(x, color='gray',dashes=(5,5), lw=2)
    
ax.axvline(0.5, color='black', lw=1)
ax.axvline(-0.5, color='black', lw=1)
ax.axhline(0, color='black', lw=1)
plt.ylabel('$p_Y(y|x_i)$', fontsize=28)
plt.xlabel(r'$y$', fontsize=28)
ax.set_frame_on(False)
plt.rc('ytick', labelsize=16)
plt.rc('xtick', labelsize=16) 
```

### Primer 2:

```python
N=0.64
sigma_n=sqrt(N)
X=array([0, 3.3])

#Izracunaj tabelo vrednosti za risanje grafa
fig, ax = plt.subplots(figsize=(12,4))
y=linspace(-2,5,1000)
for x in X:
    pnor=norm(scale=sigma_n,loc=x)
    pypx=pnor.pdf(y)
    ax.plot(y,pypx)
    ax.axvline(x, color='gray',dashes=(5,5), lw=2)
    
ax.axvline(1.5, color='black', lw=1)
ax.axhline(0, color='black', lw=1)
plt.ylabel('$p_Y(y|x_i)$', fontsize=28)
plt.xlabel(r'$y$', fontsize=28)
ax.set_frame_on(False)
plt.rc('ytick', labelsize=16)
plt.rc('xtick', labelsize=16) 
```
