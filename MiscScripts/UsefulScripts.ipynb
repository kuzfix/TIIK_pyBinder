# Get available fonts

```python
import matplotlib.font_manager

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

# List Graph styles
### Warning: Each style may only change a few parameters, therefore the order of changing styles is important!

```python
from math import sin
from math import pi
from numpy import arange
from matplotlib import pyplot as plt
x=arange(-pi,3*pi,0.01)
y=list()
for xi in x:
    y.append(sin(xi))

%matplotlib inline
for pltstyle in plt.style.available:
    print(pltstyle)
    plt.style.use('default')  # Reset all style parameters to default (I hope)
    plt.style.use(pltstyle)
    plt.plot(x,y)
    plt.title(r'Naslov z enačbo $X(y_i)=sin(\alpha+\beta-\omega*\Omega)$')
    plt.xlabel(r'To je x os $x_i$')
    plt.ylabel(r'To je y os $p_X(x)$')
    plt.show()

print('To je pa DEFAULT')
plt.style.use('default')
plt.plot(x,y)
plt.title(r'Naslov z enačbo $X(y_i)=sin(\alpha+\beta-\omega*\Omega)$')
plt.xlabel(r'To je x os $x_i$')
plt.ylabel(r'To je y os $p_X(x)$')
plt.show()

print('To je seaborn')
plt.style.use('seaborn')
plt.plot(x,y)
plt.title(r'Naslov z enačbo $X(y_i)=sin(\alpha+\beta-\omega*\Omega)$')
plt.xlabel(r'To je x os $x_i$')
plt.ylabel(r'To je y os $p_X(x)$')
plt.show()
```

# Primerjava dveh stilov

```python
from math import sin
from math import pi
from numpy import arange
from matplotlib import pyplot as plt
x=arange(-pi,3*pi,0.01)
y=list()
for xi in x:
    y.append(sin(xi))

print(plt.style.available)
%matplotlib inline
plt.style.use('default')
plt.plot(x,y)
plt.title(r'Naslov z enačbo $X(y_i)=sin(\alpha+\beta-\omega*\Omega)$')
plt.xlabel(r'To je x os $x_i$')
plt.ylabel(r'To je y os $p_X(x)$')
plt.show()

plt.style.use('fivethirtyeight')
plt.plot(x,y)
plt.title(r'Naslov z enačbo $X(y_i)=sin(\alpha+\beta-\omega*\Omega)$')
plt.xlabel(r'To je x os $x_i$')
plt.ylabel(r'To je y os $p_X(x)$')
plt.show()
```
